````python
import numpy as np
import matplotlib.pyplot as plt

# Ustawienie ziarna losowości (zapewnia powtarzalność wyników)
np.random.seed(42)

# 1. Dane do wykresu liniowego (funkcja liniowa z szumem)
x_line = np.arange(0,50)
y_line = 2 * x_line + 1 + np.random.normal(0, 1, 50)

# 2. Dane do wykresu punktowego (silna korelacja + outlier)
x_scatter = np.random.uniform(0, 100, 100)
y_scatter = 0.5 * x_scatter + np.random.normal(0, 5, 100)
# Dodanie wartości odstającej (outlier)
x_scatter = np.append(x_scatter, [99,20])
y_scatter = np.append(y_scatter, [102,80])

# 3. Dane do histogramu i boxplota (rozkład normalny)
data_dist = np.random.normal(loc=50, scale=15, size=1000)

# 4. Dane do wykresu słupkowego i kołowego (kategorie)
categories = ['Produkt A', 'Produkt B', 'Produkt C', 'Produkt D']
values_bar = [25, 45, 15, 15]
````

## 1.1. Scatter Plot (Wykres punktowy)
**Zastosowanie:** Służy do przedstawiania relacji (korelacji) między dwiema zmiennymi ilościowymi.

*   **Kiedy używać?** Gdy chcesz sprawdzić, czy wzrost jednej wartości wpływa na drugą (np. czy wyżsi ludzie ważą więcej?).
*   **Na co patrzeć?** Szukamy trendów, skupisk (klastrów) oraz punktów nietypowych (**outliers**).
*   **Funkcja:** `plt.scatter(x, y)`

## 1.2. Scatter Plot - Wskazówki customizacji

1.  **Rozmiar punktów (`s` - size):**
    *   Możesz podać jedną wartość (np. `s=100`) dla wszystkich punktów lub przekazać **wektor (tablicę)**, aby rozmiar każdego punktu zależał od innej zmiennej.
2.  **Kolor punktów (`c` - color):**
    *   Może to być pojedynczy kolor (np. `c='red'`) lub **wektor wartości**, który zostanie zmapowany na kolory (np. `c=y_scatter`).
3.  **Przezroczystość (`alpha`):**
    *   Przyjmuje wartości od `0` (niewidoczne) do `1` (pełne krycie). Bardzo przydatne, gdy punkty nakładają się na siebie (**overplotting**) – dzięki przezroczystości widać, gdzie zagęszczenie danych jest największe.
4.  **Styl markera (`marker`):**
    *   Zmienia kształt punktów. Popularne opcje: `'o'` (koło), `'x'` (krzyżyk), `'s'` (kwadrat), `'^'` (trójkąt).
5.  **Palety kolorystyczne (`cmap`):**
    *   Jeśli używasz wektora do kolorowania punktów, możesz wybrać schemat kolorów (np. `cmap='viridis'`, `cmap='plasma'`, `cmap='coolwarm'`).
6.  **Obramowanie (`edgecolors`):**
    *   Pozwala dodać obwódkę wokół punktów (np. `edgecolors='black'`), co poprawia czytelność, gdy punkty są blisko siebie.

**Dokumentacja:** [`plt.scatter()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) | [Lista dostępnych markerów](https://matplotlib.org/stable/api/markers_api.html) | [Palety kolorów (Colormaps)](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

## 2.1 Line Plot (Wykres liniowy)
**Zastosowanie:** Prezentacja zmian wartości w czasie lub wzdłuż uporządkowanej osi.

*   **Kiedy używać?** Idealny do szeregów czasowych (np. kursy walut, temperatura w ciągu dnia, wzrost sprzedaży).
*   **Teoria:** Wykres liniowy łączy punkty danych linią, co sugeruje ciągłość między nimi. Nie należy go używać do danych kategorycznych (np. marek samochodów).
*   **Funkcja:** `plt.plot(x, y)`

## 2.2 Line Plot - Wskazówki customizacji


1.  **Kolor linii (`color` lub `c`):**
    *   Możesz używać nazw kolorów (`'red'`, `'blue'`), skrótów (`'r'`, `'g'`) lub kodów HEX (`'#FF5733'`).
2.  **Styl linii (`linestyle` lub `ls`):**
    *   Definiuje wygląd linii. Popularne opcje:
        *   `'-'` (linia ciągła - domyślna)
        *   `'--'` (linia przerywana)
        *   `':'` (linia kropkowana)
        *   `'-.'` (kropka-kreska)
3.  **Grubość linii (`linewidth` lub `lw`):**
    *   Wartość liczbowa określająca szerokość linii (np. `lw=2.5`).
4.  **Znaczniki na linii (`marker`):**
    *   Pozwala wyróżnić konkretne punkty danych na linii (np. `marker='o'`, `'s'`, `'x'`).
    *   Możesz też kontrolować rozmiar znacznika (`markersize` lub `ms`) oraz jego kolor wypełnienia (`markerfacecolor` lub `mfc`).
5.  **Etykieta (`label`):**
    *   Niezbędna, jeśli chcesz, aby linia pojawiła się w legendzie po wywołaniu `plt.legend()`.

**Dokumentacja:** [`plt.plot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) | [Style linii](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html) | [Kolory w Matplotlib](https://matplotlib.org/stable/tutorials/colors/colors.html)

## 3.1 Pie Plot (Wykres kołowy)
**Zastosowanie:** Pokazuje udział poszczególnych części w całości (procenty).

*   **Kiedy używać?** Tylko gdy mamy mało kategorii (max 5-6) i ich sumaryczna wartość wynosi 100%.
*   **Uwaga projektowa:** Wykresy kołowe są często krytykowane, ponieważ ludzkie oko gorzej radzi sobie z oceną kątów i powierzchni niż z oceną długości słupków.
*   **Funkcja:** `plt.pie(sizes, labels=...)`

## 3.2 Pie Plot - Wskazówki customizacji

1.  **Etykiety (`labels`):**
    *   Lista ciągów znaków, które opisują każdy fragment koła (np. `labels=categories`).
2.  **Formatowanie procentów (`autopct`):**
    *   Pozwala wyświetlić wartości procentowe bezpośrednio na wykresie. Najczęściej używa się formatu `'%1.1f%%'` (jeden znak po przecinku).
3.  **Wyróżnienie fragmentu (`explode`):**
    *   Lista lub tablica wartości, która "wysuwa" wybrany fragment z koła (np. `explode=[0, 0.1, 0, 0]` wysunie drugi element).
4.  **Kąt startowy (`startangle`):**
    *   Określa, od którego stopnia zaczyna się rysowanie pierwszego fragmentu (często ustawia się `startangle=90`, aby zacząć od "godziny 12").
5.  **Kolory (`colors`):**
    *   Możesz przekazać listę kolorów, aby ręcznie zdefiniować wygląd każdego fragmentu.
6.  **Cień (`shadow`):**
    *   Wartość logiczna (`True`/`False`), która dodaje prosty efekt głębi pod wykresem.

**Dokumentacja:** [`plt.pie()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html) | [Przykłady wykresów kołowych](https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html)

## 4.1 Bar Plot (Wykres słupkowy) vs Histogram
Choć wyglądają podobnie, służą do zupełnie innych celów:

### Histogram
*   **Cel:** Pokazuje **rozkład** jednej zmiennej ciągłej.
*   **Działanie:** Dzieli zakres danych na przedziały (tzw. *bins*) i zlicza, ile wartości wpada do każdego z nich.
*   **Funkcja:** `plt.hist(data, bins=...)`

### Bar Plot (Wykres słupkowy)
*   **Cel:** Porównanie wartości między różnymi **kategoriami**.
*   **Działanie:** Każdy słupek reprezentuje osobną grupę (np. sprzedaż w podziale na kraje).
*   **Funkcja:** `plt.bar(x, height)`

## 4.2 Bar Plot i Histogram - Wskazówki customizacji

### Bar Plot (Wykres słupkowy) – kluczowe parametry:

1.  **Kolor słupków (`color`):**
    *   Możesz podać jeden kolor dla wszystkich lub listę kolorów (np. `color=['red', 'green', 'blue', 'yellow']`), aby wyróżnić każdą kategorię.
2.  **Obramowanie (`edgecolor`):**
    *   Dodanie koloru krawędzi (np. `edgecolor='black'`) sprawia, że słupki są wyraźniej oddzielone od siebie.
3.  **Szerokość słupków (`width`):**
    *   Wartość od 0 do 1. Domyślnie 0.8. Mniejsza wartość zwiększa odstępy między słupkami.
4.  **Wykres poziomy (`plt.barh`):**
    *   Jeśli Twoje etykiety kategorii są bardzo długie, użyj funkcji `plt.barh()` zamiast `plt.bar()`, aby słupki rosły w bok.
5.  **Grubość obramowania (`linewidth`):**
    *   Określa szerokość linii krawędzi słupka.

**Dokumentacja:** [`plt.bar()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html) | [`plt.barh()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html)

---

### Histogram – kluczowe parametry:

1.  **Liczba przedziałów (`bins`):**
    *   Najważniejszy parametr. Definiuje, na ile "koszyków" podzielone zostaną dane (np. `bins=20`). Możesz też podać konkretne granice przedziałów jako listę.
2.  **Przezroczystość (`alpha`):**
    *   Kluczowa, gdy nakładasz dwa histogramy na jeden wykres, aby widzieć, gdzie się pokrywają.
3.  **Gęstość (`density`):**
    *   Jeśli ustawisz `True`, oś Y pokaże gęstość prawdopodobieństwa zamiast liczby zliczeń (suma pól słupków wyniesie 1).
4.  **Typ histogramu (`histtype`):**
    *   Opcje takie jak `'step'` lub `'stepfilled'` pozwalają na narysowanie samej obwiedni histogramu, co bywa czytelniejsze przy wielu seriach danych.
5.  **Orientacja (`orientation`):**
    *   Możliwość ustawienia histogramu w pionie lub poziomie (`'horizontal'`).

**Dokumentacja:** [`plt.hist()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

## 5.1 Box Plot (Wykres pudełkowy)
**Zastosowanie:** Jeden z najważniejszych wykresów w analizie statystycznej. Pozwala błyskawicznie ocenić rozkład danych i ich rozrzut.

*   **Budowa:**
    *   **Pudełko:** Obejmuje 50% środkowych danych (od 1. do 3. kwartyla).
    *   **Linia w środku:** Mediana.
    *   **Wąsy:** Zasięg danych (zazwyczaj 1.5 * IQR).
    *   **Kropki poza wąsami:** Dane odstające (**outliers**).
*   **Funkcja:** `plt.boxplot(data)`

## 5.2 Box Plot - Wskazówki customizacji

1.  **Orientacja (`vert`):**
    *   Domyślnie `True` (pionowy). Ustawienie `vert=False` zmienia wykres na poziomy, co jest często czytelniejsze, gdy mamy długie nazwy kategorii.
2.  **Wypełnienie kolorem (`patch_artist`):**
    *   Domyślnie boxplot składa się tylko z linii. Jeśli ustawisz `patch_artist=True`, pudełko zostanie wypełnione kolorem, co pozwala na dalszą customizację (np. zmianę koloru wnętrza). \textcolor{red}{DODAJ WIECEJ INFO!!!}
3.  **Wcięcie (`notch`):**
    *   Ustawienie `notch=True` tworzy zwężenie wokół mediany, co pomaga wizualnie ocenić przedział ufności dla mediany.
4.  **Wartości odstające (`showfliers` oraz `flierprops`):**
    *   Możesz całkowicie ukryć wartości odstające (`showfliers=False`) lub zmienić ich wygląd za pomocą słownika parametrów `flierprops` (np. zmiana kształtu lub koloru kropek).
5.  **Pokazanie średniej (`showmeans`):**
    *   Domyślnie wykres pokazuje tylko medianę. Ustawienie `showmeans=True` nanosi na wykres dodatkowy znacznik średniej arytmetycznej.
6.  **Szerokość pudełek (`widths`):**
    *   Kontroluje, jak szerokie mają być pudełka na osi (wartość liczbowa lub lista wartości).

**Dokumentacja:** [`plt.boxplot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html) | [Opis elementów wykresu pudełkowego](https://matplotlib.org/stable/gallery/statistics/boxplot_demo.html)

### 6.1.3. Wykres skrzypcowy (`plt.violinplot`)

Wykres skrzypcowy (Violin Plot) to "młodszy i bardziej szczegółowy brat" wykresu pudełkowego. Nie tylko pokazuje kwartyle, ale też pełny kształt rozkładu (gęstość danych).

*   **Zastosowanie:** Gdy Box Plot to za mało i chcesz zobaczyć, czy dane mają np. dwa "garby" (rozkład bimodalny).
*   **Funkcja:** `plt.violinplot(data)`
*   **Kluczowe argumenty:**
    *   **`showmeans=True`**: dodaje linię oznaczającą średnią.
    *   **`showmedians=True`**: dodaje linię oznaczającą medianę.
    *   **`vert=False`**: zmienia orientację na poziomą.
